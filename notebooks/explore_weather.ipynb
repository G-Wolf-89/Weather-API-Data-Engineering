```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python

df = pd.read_csv('~/Desktop/Weather-API-DATA-Engineering/data/csv/br_final.csv') 
df2= pd.read_csv('~/Desktop/Weather-API-DATA-Engineering/data/csv/Yearly_combined.csv')
```

```python
# How have harvest outcomes changed for Minas Gerais throughout the years?
#   * Are these harvest outcomes similar to other regions in Brazil?
#   * Which region produces the most coffee (by sub-category) in Brazil?


sns.lineplot(data =df, x = 'year', y = 'million_60kgs_bag', hue= 'subdivision', marker= 'o')
plt.title('Million 60kg Bags Produced Over the Years')
plt.xlabel('Year')
plt.ylabel('Million 60kg Bags')
```
